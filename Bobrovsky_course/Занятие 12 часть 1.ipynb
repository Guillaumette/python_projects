12.5.1. Напишите программу, которая заполняет список из 100 элементов случайными числами в диапазоне от 1 до 50, и затем выводит их сумму и среднее.

```python
import random # подключим модуль работы со случайными величинами

lst = [] # создадим переменную с пустым списком
for i in range(100): # нам нужно 100 элементов
    lst.append(random.randint(1, 50)) # выведем последовательность в списке из 100 случайных элементов от 1 до 50
print("Список значений: ", lst)

# вычислим сумму
summa = 0
for i in range(0, len(lst)): 
    summa += lst[i]
print('Сумма элементов списка равна ', summa)

# вычислим среднее
mean = summa / 100 # делим сумму на количество элементов списка
print('Среднее значение элементов списка равно ', mean)
```

12.5.2. Напишите программу, которая находит в массиве введенное с клавиатуры значение и выводит его индекс (номер нахождения в массиве, начиная с нуля). Используйте break, чтобы при нахождении нужного значения не продолжать цикл.

```python
n = int(input('Число для нахождения в списке: '))

for i in range(len(lst)):
    if lst[i] == n:
        print('Индекс числа равен ', i)
        break
```

12.5.3. Напишите программу, которая находит минимальное и максимальное значения в массиве.

```python
maximum = 0 # переменная максимального значения
minimum = 1 # переменная минимального значения со значением 1, потому что список состоит из рандомных чисел от 1 до 50, а не от 0
for i in lst: # пробежимся по индексам всех элементов списка
    if maximum < lst[i]: # если наше максимальное значение, изначальное равное нулю, меньше элемента списка,
        maximum = lst[i] # обновляем значение максимума до значения этого элемента и на выходе получим максимум
    if minimum > lst[i]: # присвоив минимально возможное значение этой переменной, проверяем, есть ли большие, чтобы получить корректный минимум
        minimum = lst[i] # если большие есть, обновляем значение минимума
print('Максимальное значение в массиве равно', maximum)
print('Минимальное значение в массиве равно', minimum)
```
