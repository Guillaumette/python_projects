13.4.1. Напишите программу, которая с помощью вложенных циклов вводит с клавиатуры матрицу переменного размера (размер также вводится с клавиатуры).

```python
import random # импортируем библиотеку для работы со случайными числовыми значениями

n = int(input('Количество строк: ')) 
m = int(input('Количество столбцов: ')) 

lst = [] # основной список
for i in range(n): # цикл по строкам 
    lst2 = [] # вложенный подсписок
    for j in range(m): # цикл по столбцам
        lst2.append(random.randint(1, 9)) # заполняем подсписок значениями от 1 до 9
    lst.append(lst2) # добавляем в основной список новый подсписок

print('Матрица переменного размера:', lst)
```

13.4.2. Напишите программу, которая проверяет, равны ли два трёхмерных массива. Сначала сравните их размеры по каждому измерению, и только если они совпадают, сравните массивы по значениям. <br>
<i> Когда будет выяснено, что массивы не равны, полностью прервите дальнейшие расчёты. В заключение, за пределами всех циклов, выводите сообщение, равны массивы или нет. </i> <br>
Подразумевается, что "внутри" каждого измерения все элементы массива одинаковой длины. То есть в таблице не может быть строк или столбцов разной длины: массивы -- всегда "правильные" кубы.

```python
import random # импортируем библиотеку для работы со случайными числовыми значениями

# формируем два трехмерных массива
lst1 = [] # первый
lst2 = [] # второй
for i in range(3): 
    lst3 = [] # вложенный подсписок для первого основного
    lst4 = [] # вложенный подсписок для второго основного
    for j in range(3):
        # заполняем их значениями
        lst3.append(random.randint(1, 9))
        lst4.append(random.randint(1, 9))
    # и заполняем основные списки
    lst1.append(lst3) 
    lst2.append(lst4)

print('Первый список:', lst1)
print('Второй список:', lst2)

# Сравним длины списков
if len(lst1) == len(lst2):
    flag = True
else:
    flag = False
if flag:
    print('Списки равны по длине')
else:
    print('Списки не равны по длине')

# Сравним длины подсписков
flag = True
for i in range(len(lst1)):
    for j in range(len(lst2)):
        if len(lst1[i]) != len(lst2[j]):
            flag = False
            break
        
if flag:
    print('Списки равны по размеру подсписков')
else:
    print('Списки не равны по размеру подсписков')
    
# Сравним равенство значений в списках
for i in range(3):
    for j in range(3):
        if lst1[i][j] == lst2[i][j]:
            flag = True
        else:
            flag = False
if flag:
    print('Массивы равны по значениям')
else:
    print('Массивы не равны по значениям')
```
