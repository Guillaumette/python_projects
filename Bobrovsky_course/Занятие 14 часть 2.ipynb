14.4.4. Запишите в переменную, содержащую строку, её саму, перевёрнутую от конца к началу.

```python
s = input('') # создадим строку
lst = list(s) # конвертируем ее в список для доступа к отдельным элементам

for i in range(len(s) // 2): # делим длину строки, чтобы операция смены мест каждой пары элементов произошла только один раз
        a = lst[i] # создадим временную переменную для хранения элемента списка
        lst[i] = lst[len(s) - i - 1] # теперь переносим последний элемент исходной строки в начало
        lst[len(s) - i - 1] = a # меняем местами первый и последний элементы, а потом симметрично 2 с начала и 2 с конца и т.д.
        s2 = ''.join(lst) # добавляем в новую переменную строку, но уже с первым элементом в конце, а последним в начале и т.д.
        s = s2
print(s)
```

14.4.5. Введите с клавиатуры две строки, и создайте третью строку, которая будет равна сумме двух этих строк (двум строкам, сцепленным друг за другом). Не используйте стандартную операцию сцепления строк "+".

```python
# создадим строки
s1 = input('')
s2 = input('')
# преобразуем их в списки
lst1 = list(s1)
lst2 = list(s2)
# введем переменную-список для записи в нее элементов
sum_strings = []

# положим в эту переменную все элементы, пройдясь по двум спискам
for i in range(len(lst1)):
    sum_strings.append(lst1[i])

for i in range(len(lst2)):
    sum_strings.append(lst2[i])

# преобразуем ее в строку
s3 = ''.join(sum_strings)
print(s3)
```

14.4.6. Напишите программу, которая выделяет из строки подстроку -- начиная с заданной позиции и заданной длины, и формирует из неё отдельную строку.



```python
# Введем необходимые переменные
s = input('Введите строку: ')
position = int(input('Введите значение позиции: '))
length = int(input('Введите значение длины подстроки: '))
# Конвертируем строку в список для работы с отдельными элементами
lst = list(s)
# создадим переменную для списка-новой подстроки
lst2 = []

# защита от дурака: если пользователь вводит пустую строку либо отрицательное значение позиции или длины подстроки;
# также если сумма позиции минус 1 (то есть индекса первого элемента) и длины строки больше длины строки, 
# то выводится сообщение о невозможности выполнить операцию
if len(lst) < 1 or position < 1 or length < 1 or position > len(lst) or length > len(lst) or position - 1 + length > len(lst):
    print('Невозможно выполнить операцию со строкой')
#  в ином случае, если все нормально, то идем по списку в границах от позиции минус 1(индекс первого элемента)
# до длины подстроки плюс значение первого индекса (чтобы код работал исправно вне зависимости от заданных координат)
else:
    for i in range(position - 1, length + position -1):
        lst2.append(lst[i]) # добавляем элементы в новый список
        s2 = ''.join(lst2) # конвертируем список в новую строку
    print(s2)




```
