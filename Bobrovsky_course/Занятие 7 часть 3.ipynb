Все дома сделаны по типовому проекту -- размер каждой стены W метров в ширину по одной стороне, L метров в ширину по другой стороне, и H метров в высоту. На покраску случайным образом поступают банки одного из двух типов -- объёмом, достаточным для покраски 15 кв.м. и объёмом на 42 кв.м. Заранее неизвестно, какой вид краски прибудет.

Имеются небезосновательные подозрения, что существенная часть краски сбывается налево.
Ваша задача как общественного инспектора проверить, сколько минимально банок краски требуется для конкретной модели дома.

Входные данные, четыре числа:
объём банки краски (15 или 42 или любое другое)
размеры стены (W L H)

```python
w = int(input()) # ширина одной стороны
l = int(input()) # ширина другой стороны
h = int(input()) # высота
v = int(input()) # объем банки краски

# рассчитаем площадь всех четырех стен, которые подлежат покраске
square = w * h * 2 + l * h * 2

# Посчитаем минимальное количество банок краски и прибавим единицу в случае вещественного числа
print('Минимально', square // v + 1, 'банок требуется для покраски')
```

7.9. <br>  Напишите программу, которая вычисляет возраст пёсика как если бы он был человеком. По "человечески" собачий возраст считается так: первые два года реального возраста пса идут по 10.5 человечьих лет каждый, а каждый следующий -- по 4 человеческих года.

```python
# Введем переменную, обозначающую реальный возраст собаки
real_age = int(input())

# Для одного реального года и двух реальных лет нужно считать отдельно
if real_age == 1:
    dog_age = 10.5
elif real_age == 2:
    dog_age = 10.5 * 2

# А дальше приводим к формуле, учитывая особенные первые два года собаки, потом вычитая их из реального возраста 
# и домножая на 4 человечческих года
elif real_age > 2:
    dog_age = 10.5 * 2 + (real_age - 2) * 4
print(dog_age, 'человеческих год(а)')
```
