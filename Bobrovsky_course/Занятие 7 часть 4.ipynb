7.10. <br> Оксана вычитала в астрологическом журнале способ проверки совместимости с парнем. Оказывается, если сумма всех цифр её года рождения больше суммы всех цифр года рождения молодого человека, то он ей подходит. <br>
Помогите Оксане отсеивать неподходящих парней! <br>
Года рождения рассматриваем в диапазоне от 1900 до 2100. <br>
Каждый год вводится одним числом, не отдельными цифрами.

```python
year = int(input())

# Найдем сумму всех цифр: сначала разделим год на 1000, вытащив первую цифру, потом разделим год на 100, 
# а от получившегося числа возьмем остаток - это вторая цифра, далее таким же образом отделим третью цифру, 
# поделив сначала год на 10, а потом отделив от этого числа остаток, и наконец, 
# возьмем остаток от деления года на 10 - это четвертая цифра

summa = year // 1000 + ((year // 100) % 10) + ((year // 10) % 10) + (year % 10)

print('Сумма равна', summa)
```
