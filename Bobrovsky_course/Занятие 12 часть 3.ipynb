12.5.7. Напишите программу, которая печатает элементы, одновременно присутствующие в двух массивах (предварительно заполните их случайными числами). Используйте вложенные циклы.

```python
import random # подключим модуль работы со случайными величинами

lst1 = [] # создадим переменную с пустым списком
for i in range(20): # сделаем 20 элементов
    lst1.append(random.randint(1, 50)) # выведем последовательность в списке из 20 случайных элементов от 1 до 50
print('Список значений первого списка:', lst1)

lst2 = [] # создадим переменную с пустым списком
for i in range(20): # сделаем 20 элементов
    lst2.append(random.randint(1, 50)) # выведем последовательность в списке из 20 случайных элементов от 1 до 50
print('Список значений второго списка:', lst2)

print('Элементы, присутствующие в двух списках:')
for i in range(len(lst1)):
    for j in range(len(lst2)):
        if lst1[i] == lst2[j]:
              print(lst1[i], end = " ")
```

12.5.8. Напишите программу, которая сортирует массив из 10 элементов по возрастанию.

```python
import random # подключим модуль работы со случайными величинами

lst = [] # создадим переменную с пустым списком
for i in range(10): # нам нужно 10 элементов
    lst.append(random.randint(1, 50)) # выведем последовательность в списке из 10 случайных элементов от 1 до 50
print('Список значений:', lst)

for j in range(len(lst) - 1): # вводим переменную j для того, чтобы цикл прошелся по списку столько раз, сколько необходимо, чтобы отсортировались все элементы по возрастанию
    for i in range(len(lst) - 1 - j): # бежит по индексам списка на длину меньше длины списка на 1 и на j, чтобы не словить list index out of range
        if lst[i] > lst[i + 1]: # если элемент слева больше соседа справа,
            lst[i], lst[i + 1] = lst[i + 1], lst[i] # меняем их местами
print('Отсортированный список значений:', lst)
```

12.5.9. Реализуйте логику метания кубиков из игры Dungeons & Dragons. Логика кубика записывается формулой NdM, где N -- количество кубиков, а M -- количество граней каждого кубика. Например, 3d12 означает бросок трёх двенадцатигранных кубиков. Соответственно, минимально может выпасть 3, а максимально 36. <br>
Вам надо вывести результат метания кубика, для которого введены N и M. <br>
<i> Не используйте в этой задачке массивы, это задача на случайные значения. </i>

Тут надо смоделировать сам процесс: бросаем N раз кубик с M гранями, и суммируем выпавшие значения.

```python
import random # подключим модуль работы со случайными величинами

n = int(input('Количество бросков: '))
m = int(input('Количество граней одного кубика: '))
sum = 0 # переменная-счетчик для суммы выпавших значений

for i in range(n): # запускаем цикл по количеству бросков
    a = random.randint(1, m) # зададим границы возможного значения переменной а от единицы до m
    print('Результат одного броска:', a)
    sum += a # суммируем все выпавшие значения а и получим результат бросания кубиков
print('Результат метания кубика:', sum)
```
