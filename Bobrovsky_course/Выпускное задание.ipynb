Выпускное задание:

Напишите функцию с двумя параметрами-строками. Функция возвращает булево значение True если вторая строка содержится в первой строке как подстрока, и False в противном случае.
Например,

f("12345", "234") = True

f("12345", "235") = False

Для решения стандартные возможности типа s1 in s2 или слайсы использовать нельзя, только условиями и циклами. Допускается использовать только то, что проходили на начальном курсе.

```python
def check_substring(s1, s2):
    # Первая строка не может быть подстрокой второй строки
    if len(s1) <= len(s2):
        return False
    # flag нужен, чтобы найти первый совпадающий элемент
    flag = True
    # index нужен для итерирования по элементам возможной подстроки
    index = 1
    # Берем элемент первой строки
    for i in s1:
        if flag:
            # пока не найдем элемент первой строки, равный эл-ту второй строки, не меняем флаг
            for j in s2:
                if i == j:
                    # как только нашли эл-т 1 строки, равный эл-ту 2 строки, меняем флаг и смотрим следующий эл-т 1 строки
                    flag = False
                    break
                else:
                    break
        # приходим сюда, когда флаг = True, что означает, что мы нашли 1-ый эл-т подстроки в проверочной (1) строке
        else:
            # пока индекс плюс 1 меньше длины подстроки,
            while index + 1 < len(s2):
                # начинаем проверку со 2 эл-та 2 строки, потому что первый уже равен какому-то эл-ту 1 строки
                # также указываем условие, когда индекс равен длине строки, чтобы не исключить последний элемент
                if index == len(s2) and s2[index] == i:
                    return True
                elif s2[index] == i:
                    #  также пока эл-ты строк совпадают, прибавляем индекс, чтобы проверить след. эл-т 2 строки на совпад-е
                    index += 1
                else:
                    return False
                # если какой-то эл-т 2 строки с послед. эл-том 1 строки, возвращаем False, потому что s2 не входит в s1
            else:
                return True


str1 = '12345'
str2 = '345'
check_substring(str1, str2)
```
