11.3.4. Удалите из числа, состоящего из двух или более разрядов, заданную цифру. Например, если из числа 1203453 удалить цифру 3, то получится число 12045.
Использовать можно только арифметические операции.

Подсказка: используйте идею из задачи 9.3.5, когда из 6 цифр билета выделялось по одной цифре справа.
Идея, что сперва выделяете правую цифру числа, и одновременно можно конструировать результат. Только нужен коэффициент, который принимает значения 1,10,100,... -- на него умножаем текущую выделенную цифру и прибавляем к накопителю-результату.

```python
num = int(input("Введите число: "))
del_symbol = int(input('Введите число для удаления: '))
coef = 1
result = 0

if num // 10 == 0: # защита от дурака, если пользователь решит ввести число меньше, чем двухразрядное
  print("Введено некорректное число")

while num > 0:
  x = num % 10 # отделяем правую цифру
  if x != del_symbol: # если отделенная цифра справа не является числом для удаления,
    y = x * coef # домножаем на коэффициент
    coef *= 10 # домножаем коэффициент на 10, чтобы следующее число соответствовало своему разряду в изначальном
    result += y # прибавляем получившееся число к результату
  num //= 10 # удаляем отделенную цифру из основного числа
  
print(result)
```

Или другой способ:

```python
num = int(input("Введите число: "))
del_symbol = int(input('Введите число для удаления: '))
result = "" # типа строка, потому что конкатенировать друг с другом можно только строки

if num // 10 == 0: # защита от дурака, если пользователь решит ввести число меньше, чем двухразрядное
  print("Введено некорректное число")

while num > 0:
  y = num % 10 # отделяем по одной цифре справа
  if y != del_symbol: # если отделенная цифра справа не является числом для удаления,
    result = str(y) + result # то преобразуем ее в строку и прибавим к результату, который тоже является строкой,
  num //= 10 # и удалим отделенную цифру из основного числа

result = int(result) # преобразуем результат обратно в целое число
print(result)
```

11.3.5. Винни-Пух готовится к зиме и решил не лакомиться медом, а относить полный горшочек объемом X горстей, и перекладывать мёд в бочку. В первый день он донёс горшочек полным, но начиная со следующего дня он больше не смог бороться с искушением и начал выедать из горшочка Y горстей. Причём на каждый следующий день он выедал на Y горстей больше, чем в предыдущий день.

На входе программы два числа: X -- объём горшочка, Y -- выедаемые горсти. Посчитайте, сколько всего мёда Винни-Пух накопит в бочке, пока горшочек не опустеет, и сколько на это уйдёт дней.

```python
x = int(input('Объем горшка: '))
y = int(input('Съеденные горсти: '))
days = 0 # количество дней
honey = x # количество донесенного меда в первый день

while x > 0: # пока горшок не опустел,
  x -= y # каждый день отнимаем по количеству заданных горстей,
  days += 1 # прибавляем счетчик дней,
  honey += x # прибавляем количество накопленного меда
print('Винни-Пух накопит в бочке', honey, 'горсти(-ей), и на это уйдет', days, 'дня(-ей)')
```
