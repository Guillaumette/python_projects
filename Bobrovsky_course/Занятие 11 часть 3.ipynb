11.3.6. Проверяем гипотезу Коллатца -- простейшую из нерешённых задач математики.
Возьмём любое целое неотрицательное число. Если оно чётное, поделите его на два. Если нечётное, умножьте на три и прибавьте один. Утверждается, что если такой цикл повторять достаточно долго, то рано или поздно в результате мы получим единицу.
Напишите программу, которая получает на вход целое неотрицательное число, и затем выполняет бесконечный цикл, выполняя над текущим значением вышеприведённые манипуляции. Когда в итоге будет получено значение 1, завершите цикл и выведите количество итераций.

```python
n = int(input('Неотрицательное число: '))
it = 0 # количество итераций

if n < 0: # исключим возможность введения неподходящего числа
    print('Введено некорректное число') 
    
while n != 1: # пока n не равно единице,
    if n % 2 == 0: # если число четное,
        n = n / 2 # делим его на два
    else: # в противном случае (если число нечетное),
        n = n * 3 + 1 # умножим n на 3 и прибавим 1
    it += 1 # прибавим к счетчику итераций единицу
    
print('Количество итераций равно', it)
```

11.3.7.(дополнительное). Определите, сколько единичных разрядов (сколько разрядов с цифрой 1, разряды с цифрой 0 не считаются) в двоичном представлении вводимого числа N. <br>
Подсказка: само число переводить в двоичный вид необязательно.

```python
n = int(input('Число для перевода в двоичное представление: '))
one = 0 # счетчик единиц
x = '' # переменная для записи числа в двоичной системе в формате строки
holdn = n # сохраним переменную n для работы со вторым циклом подсчета единиц

# Переведем число в двоичную систему
while n > 0: # пока n больше нуля,
    x = str(n % 2) + x # возьмем остаток от него, преобразуем в строку и запишем в x
    n //= 2 # и разделим нацело на 2
    
while holdn != 1: # пока n не равно 1,
    a = holdn // 2 # разделим нацело на 2,
    b = holdn - a * 2 # вычтем остаток от деления, умноженный на 2, от изначального числа,
    holdn = a # обновим значение n
    if a == 1 or b == 1: # если деление нацело или операция от вычитания остатка * 2, равны единице,
        one += 1 # прибавим это в счетчик единиц

print('Число в двоичном представлении равно', x)
print('Количество единичных разрядов числа в двоичном представлении равно', one)
```
