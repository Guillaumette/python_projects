13.4.5. Напишите программу, которая рассчитывает сумму элементов каждой строки и каждого столбца матрицы, затем складывает эти суммы по горизонтали и вертикали, и проверяет, что две итоговые суммы равны (проверка бухгалтерского баланса). <br>
<i> Заводим два дополнительных массива, в первый записываем значение сумм по каждой строке, во второй -- значение сумм по каждому столбцу. Выводим эти одномерные массивы, и в заключение суммируем их и проверяем две суммы на равенство.
Исходный массив делаем таким, что количество строк и столбцов разное. 
Длины всех строк и длины всех столбцов считаем одинаковыми (то есть матрица -- ровный прямоугольник). <i> 

```python
matrix = [[1, 2, 4, 29], # создаем матрицу с разным количеством строк и столбцов
[3, 4, 6, 1]]

sum_strings = [] # массив для суммы элементов строк
sum_columns = [] # массив для суммы элементов столбцов

# посчитаем сумму элементов строк
for i in range(len(matrix)): # индекс подсписка
    strings = 0 # переменная для хранения суммы строк
    for j in range(len(matrix[0])): # индекс элемента подсписка, указываем, что идем с первого элемента и указываем на кол-во строк
        strings += matrix[i][j] # складываем элементы, записываем в переменную
    sum_strings.append(strings) # добавляем ее в массив для суммы элементов строк
        
print('Сумма строк:', sum_strings)

# посчитаем сумму элементов столбцов
for i in range(len(matrix[0])): # начинаем с первого столбца
    columns = 0 # переменная для хранения суммы столбцов
    for j in range(len(matrix)): # идем по количеству столбцов
        columns += matrix[j][i] # складываем элементы, записываем в переменную
    sum_columns.append(columns) # добавляем ее в массив для суммы элементов столбцов
    
print('Сумма столбцов:',sum_columns)

# Суммируем получившиеся массивы с суммами строк и столбцов по горизонтали и вертикали
sum_horizon_strings = 0 # переменная суммы строк по горизонтали
sum_horizon_columns = 0 # переменная суммы столбцов по горизонтали

# по горизонтали сумма строк
for i in sum_strings:
    sum_horizon_strings += i

# по горизонтали сумма столбцов
for i in sum_columns:
    sum_horizon_columns += i
    
print('Сумма строк:', sum_horizon_strings, 'сумма столбцов:', sum_horizon_columns)

# проверяем на равенство
if sum_horizon_strings == sum_horizon_columns:
    print('Суммы строк и столбцов равны')
else:
    print('Суммы строк и столбцов не равны')
```

13.4.6. Напишите программу, которая транспонирует массив (когда строки становятся столбцами, а столбцы строками).<br>
<i> Исходный массив делаем таким, что количество строк и столбцов разное.</i> <br>
Результат записывайте, конечно, во второй массив нужной размерности.

```python
matrix = [[1, 2, 4, 29], # создаем матрицу с разным количеством строк и столбцов
[3, 4, 6, 1]]

new_matrix = [] # создаем новый массив для записи результата
for i in range(len(matrix[0])): # идем по первому подсписку матрицы (итерируемся по строке)
    string = [] # создаем промежуточную переменную для хранения каждого столбца исходной матрицы, который становится строкой
    for j in range(len(matrix)): # переменная для элементов подсписков исходной матрицы (итерируемся по столбцу)
        string.append(matrix[j][i]) # переворачиваем, как нам нужно
    new_matrix.append(string) # записываем в матрицу для записи результата

print('Транспортированная матрица:', new_matrix)
```
