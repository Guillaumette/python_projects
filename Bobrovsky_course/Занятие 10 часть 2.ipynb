10.5.3. Напишите программу, которая печатает такой узор:

ooooooooooooooooo                                                       
ooooooooooooooooo                                                       
ooooooooooooooooo                                                       
oooo                                                                    
oooo                                                                    
oooo                                                                    
ooooooooooooooooo                                                       
ooooooooooooooooo                                                       
ooooooooooooooooo                                                       
             oooo                                                       
             oooo                                                       
             oooo                                                       
ooooooooooooooooo                                                       
ooooooooooooooooo                                                       
ooooooooooooooooo 

```python
for i in range(16): # цикл по количеству строк
    for j in range(17): # цикл по количеству символов на строке / количеству столбцов
        if 0 < i < 4 or 6 < i < 10 or 12 < i < 16: # рисуем три длинные горизонтали
            print("o", end = "")
        if 3 < i < 7 and j < 4: # рисуем три короткие горизонтали слева
            print("o", end = "")
        if 9 < i < 13: # рисуем в два этапа три короткие горизонтали справа
            if 0 < j < 14: # сначала рисуем нужное количество пробелов от левого края
                print(" ", end = "")
            if j >= 13: # и рисуем оставшиеся символы трех горизонталей
                print("o", end = "")
    print("")
```

10.5.4. Напишите программу, которая выводит таблицу умножения от 1 до 9 в формате

1x1 = 1 <br>
1x2 = 2 <br>
1x3 = 3 <br>
<br>
... <br>
5x1 = 5 <br>
5x2 = 10 <br>
5x3 = 15 <br>
... <br>
9x1 = 9 <br>
9x2 = 18 <br>
9x3 = 27

```python
print('Таблица умножения')
print('')
for i in range(1, 10):
    for j in range(1, 10):
        print(i, "x", j, "=", i * j)
    print('')
```
