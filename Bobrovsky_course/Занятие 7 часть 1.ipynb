7.1. <br> В вашем отделе работают три программиста -- джун, миддл и сеньор. При этом не обязательно зарплата сеньора самая высокая. Введите в программу три числа (зарплаты) и определите, у кого она самая высокая, у кого самая низкая, и во сколько раз самая большая отличается от самой маленькой.

```python
# Введем три переменные, обозначающие зарплаты джуна, миддла и сеньора
jun = 85000
middle = 120000
senior = 200000

# Определим, у кого самая высокая зарплата
if jun > middle and jun > senior:
    print('У джуна самая высокая зарплата')
elif middle > jun and middle > senior:
    print('У миддла самая высокая зарплата')
elif senior > jun and senior > middle:
    print('У сеньора самая высокая зарплата')

# Определим, у кого самая низкая зарплата
if jun < middle and jun < senior:
    print('У джуна самая низкая зарплата')
elif middle < jun and middle < senior:
    print('У миддла самая низкая зарплата')
elif senior < jun and senior < middle:
    print('У сеньора самая низкая зарплата')

# Посчитаем разницу между самой высокой и самой низкой зарплатами
difference = senior / jun
print('Зарплата сеньора отличается от зарплаты джуна в', difference, 'раз')
```

Не используйте никакие стандартные функции.

```python
# Введем переменные с двумя вещественными чисами
x = 2.657489
y = 2.657375

# Так как при совпадении до третьего знака после запятой включительно остаток деления одного числа на другое 
# должен быть равен не меньше 0.0001, напишем программу со следующим условием:
if x % y >= 0.0001:
    print('Числа совпадают с точностью до третьего знака после запятой включительно')
else:
    print('Числа не совпадают с точностью до третьего знака после запятой включительно')
```

7.3. <br> Введите в программу свой год рождения, и напишите код, который определит, високосный ли это год.

```python
# Введем переменную, обозначающую год моего рождения
birth_year = 1994

# Високосный год должен отвечать ряду условий: делиться нацело на 4, не делиться без остатка на сто 
# либо делиться без остатка на 400. Введя все эти условия, получим верный результат
if birth_year % 4 == 0 and birth_year % 100 != 0 or birth_year % 400 == 0:
    print(birth_year, 'год является високосным')
else:
    print(birth_year, 'год не является високосным')
```

7.4. <br> Введите целое число и напечатайте, чётное ли оно.

```python
# Вводим переменную
number = int(input())

# Если остаток от деления числа на 2 равен нулю, оно будет четным, если нет - нечетным.
if number % 2 == 0:
    print(number, '-- четное число')
else:
    print(number, '-- нечетное число')
```
