12.5.4. Напишите программу, которая находит второй по величине (второй максимальный) элемент массива. <br>
<i> Если исходный массив упорядочить по убыванию, то первое максимальное будет его первым (нулевым) элементом, второе следующим, независимо от их значений (равны они или нет). Однако для решения сортировку массива не делайте, используйте только один цикл </i>.

```python
import random # подключим модуль работы со случайными величинами

list100 = [] # создадим переменную с пустым списком
for i in range(100): # нам нужно 100 элементов
    list100.append(random.randint(1, 50)) # выведем последовательность в списке из 100 случайных элементов от 1 до 50
print('Список значений:', list100)

n = 100 # длина массива

# Найдем оба максимума
for i in range(n):    
    if i == 0: # начальное значение максимума
        first_max100 = list100[i] # вводим переменную первого максимума и присваиваем ей значение первого элемента
        second_max100 = 0 # вводим переменную второго максимума со значением вне пределов списка
    elif list100[i] > first_max100: # иначе, если элемент больше первого максимума,
        second_max100 = first_max100 # сначала второй максимум получает значение этого элемента,
        first_max100 = list100[i] # а первый идет дальше
        
    # отдельным условием прописываем, чтобы второй максимум был больше всех элементов списка, но меньше первого максимума
    if list100[i] > second_max100 and list100[i] < first_max100: 
        second_max100 = list100[i]

print('Первый максимум равен', first_max100)  
print('Второй максимум равен', second_max100)
```
